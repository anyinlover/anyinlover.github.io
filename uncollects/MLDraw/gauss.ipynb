```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Example with different parameters (multiple plots)
mu1 = 0
sigma1 = 1

mu2 = 2
sigma2 = 0.5

x = np.linspace(-5, 7, 200)  # Wider range to accommodate both curves

y1 = stats.norm.pdf(x, mu1, sigma1)
y2 = stats.norm.pdf(x, mu2, sigma2)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=f"μ={mu1}, σ={sigma1}")
plt.plot(x, y2, label=f"μ={mu2}, σ={sigma2}")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Gaussian Distributions (Multiple Plots)")
plt.grid(True)
plt.legend()
plt.savefig("multiple_gaussians.png", dpi=300, bbox_inches="tight")
plt.show()
```
