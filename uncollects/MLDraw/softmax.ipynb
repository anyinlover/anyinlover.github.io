```python
import numpy as np
import matplotlib.pyplot as plt


# Softmax function
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Subtract max for numerical stability
    return exp_x / exp_x.sum(axis=0)


# Generate a range of logits (input to softmax)
x = np.linspace(-10, 10, 100)

# Apply softmax to each value in x (as a scalar input)
y = softmax(x)

# Plot the result
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Softmax Function", color="b")
plt.title("Softmax Function")
plt.xlabel("Input (logits)")
plt.ylabel("Probability")
plt.grid(True)
plt.legend()
plt.savefig("softmax.png", dpi=300, bbox_inches="tight")
plt.show()
```
