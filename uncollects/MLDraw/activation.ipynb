```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

# Define the activation functions


def relu(x):
    return np.maximum(0, x)


def gelu(x):
    return 0.5 * x * (1 + erf(x / np.sqrt(2)))


def silu(x):
    return x * sigmoid(x)


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def tanh(x):
    return np.tanh(x)


def swiglu(x):
    return x * sigmoid(x)


# Define the input range
x = np.linspace(-5, 5, 400)

# Compute activations
relu_vals = relu(x)
gelu_vals = gelu(x)
silu_vals = silu(x)
sigmoid_vals = sigmoid(x)
tanh_vals = tanh(x)
swiglu_vals = swiglu(x)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot each activation function
plt.plot(x, relu_vals, label="ReLU", color="blue")
plt.plot(x, gelu_vals, label="GeLU", color="green")
plt.plot(x, silu_vals, label="SILU (Swish)", color="orange")
plt.plot(x, sigmoid_vals, label="Sigmoid", color="purple")
plt.plot(x, tanh_vals, label="Tanh", color="brown")
# plt.plot(x, swiglu_vals, label="SwiGLU", color="red")

# Add labels and title
plt.title("Comparison of Activation Functions")
plt.xlabel("x")
plt.ylabel("Activation Output")
plt.legend()

# Display the plot
plt.grid(True)
plt.savefig("activations.png", dpi=300, bbox_inches="tight")
plt.show()
```
